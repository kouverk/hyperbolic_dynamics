```python
import matplotlib.pyplot as plt
import numpy as np


def iterate(x):
    x = x**2 - 3/2
    return x 

arr = []
x = 0 
for i in range(10000):
    x = iterate(x)
    arr.append(x)

data = arr
plt.figure(figsize=(22, 15))  


color_sequence = ['r', 'g', 'b', 'y', 'm', 'k']  # Example color sequence
section_length = 100
num_sections = len(data) // section_length

# Plot the data in sections with different colors
for i in range(num_sections):
    start_index = i * section_length
    end_index = (i + 1) * section_length
    color = color_sequence[i % len(color_sequence)]  # Repeat color sequence
    plt.plot(range(start_index, end_index), data[start_index:end_index], 'o', color=color, markersize=3)

# Set plot title, labels, and grid
plt.title('Different Colored Sections of Data')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)

plt.show()
```

```python
def has_duplicates(arr):
    seen = {}
    for elem in arr:
        if elem in seen:
            return True
        seen[elem] = True
    return False

# Example usage:
my_array = data

if has_duplicates(my_array):
    print("Array contains duplicate values.")
else:
    print("Array does not contain duplicate values.")
```
